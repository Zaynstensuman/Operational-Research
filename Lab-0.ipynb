My favorite function is **factorial function** .   
I like the factorial function because it appears in **many areas of mathematics**, such as:
- probability and statistics, combinations, growth patterns and  algorithms




```python
# Example: Length of a string
def factorial(n):
    """Return the factorial of n."""
    result = 1
    for i in range(1, n + 1):
        result *= i
    return result

# Compute a few interesting values
print("3! =", factorial(3))
print("5! =", factorial(5))
print("7! =", factorial(7))
print("10! =", factorial(10))
```

## Creating a List of Factorial Values

The **factorial function**, I decided to create my list in a creative way by
computing the factorials of the numbers from 1 through 10. This gives me **10 values**, each one growing
much larger than the previous because factorial values increase very quickly.

Then I also generated the list by calling my factorial function for each number from 1 to 10 and collecting the results
into a Python list. Then, using Pythonâ€™s built-in functions, I computed the **sum**, **minimum**, **maximum**,
and the **average** of these factorial values.

```python
# List of factorial values from 1! to 10!
factorial_list = [factorial(n) for n in range(1, 11)]
print("Factorial List:", factorial_list)

# Required calculations
print("Sum:", sum(factorial_list))
print("Minimum:", min(factorial_list))
print("Maximum:", max(factorial_list))
print("Average:", sum(factorial_list) / len(factorial_list))
```
