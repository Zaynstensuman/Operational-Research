## 1) Factorials — How many zeros does `1000!` end with?

Trailing zeros come from factors of 10 (pairs of 2 and 5). Count the exponent of 5 in `1000!`:

```python

n = 1000
count = 0
p = 5
while p <= n:
    count += n // p
    p *= 5
count
```

**Result:** `249` trailing zeros.

## 2) Big Numbers — Sum of digits of `100**13 - 26`

Observation: `100**13 = 10**26`. So the number is `10**26 - 26`. Let's compute the digit sum:

```python

num = 100**13 - 26
digit_sum = sum(int(d) for d in str(num))
num, digit_sum
```

## 3) Handshakes — Two soccer teams of 11 players; winning team also shakes hands with each other

Cross-team handshakes = `11 × 11 = 121`. Internal handshakes among 11 players = `C(11,2) = 55`. Total = 176.

```python

players_per_team = 11
cross_team = players_per_team * players_per_team  # 11*11 = 121
winning_internal = players_per_team * (players_per_team - 1) // 2  # C(11,2) = 55
total = cross_team + winning_internal
cross_team, winning_internal, total
```
